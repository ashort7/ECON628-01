```python
import matplotlib.pyplot as plt
%matplotlib inline

TechComp = ['Avrg_AAPL', 'Avrg_GOOG']

for comp in TechComp:
    fig = plt.figure(figsize=(5, 5))
    plt.hist(df['{}'.format(comp)], bins=15, color='g', label='{} Stock Price'.format(comp))
    plt.ylabel("Frequency")
    plt.xlabel("Stock Price")
    plt.title("{} Stock Price".format(comp))
    plt.show()

```
